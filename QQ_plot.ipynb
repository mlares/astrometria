```python
import numpy as np
import statsmodels.api as sm
import pylab
import stats
from scipy import stats
import pylab
from matplotlib import pyplot as plt
```

```python
test = np.random.weibull(1, 1000)
sm.qqplot(test, line='45')
pylab.show()
```

```python
N = stats.norm(loc=3, scale=2)
x = N.rvs(size=100)

fig = plt.figure(figsize=(10,6))
stats.probplot(x, dist="norm", plot=pylab)
pylab.show()
```

```python
def qq_plot_norm(loc=0, scale=1, size=100):

    N = stats.norm(loc=loc, scale=scale)
    x = N.rvs(size=100)

    fig = plt.figure(figsize=(10,6))
    stats.probplot(x, dist="norm", plot=pylab)
    pylab.show()
```

```python
qq_plot_norm(0, 10, 100)
```

```python
(qt, qe), (slope, intercept, r) = stats.probplot(x, dist="norm")
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot()
ax.axline((0, intercept), slope=slope, color='coral')
ax.plot(qt, qe, marker='o', linestyle='None')
ax.grid()
```

```python
def qq_norm(loc=0, scale=1, size=100, ref='norm'):
    N = stats.norm(loc=loc, scale=scale)
    x = N.rvs(size=100)
    (qt, qe), (slope, intercept, r) = stats.probplot(x, dist=ref)
    return (qt, qe), (slope, intercept, r)
```

```python
res1 = qq_norm(loc=0, scale=1, size=50, ref='norm')
(qt1, qe1), (slope1, intercept1, r1) = res1
```

```python
res2 = qq_norm(loc=0, scale=2, size=50, ref='norm')
(qt2, qe2), (slope2, intercept2, r2) = res2
```

```python
opts = {'marker': 'o', 'linestyle': 'None'}

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot()
ax.plot(qt1, qe1, **opts)
ax.plot(qt2, qe2, **opts)
```

```python
def qq_norm_rvs(rvs, ref='norm'):
    (qt, qe), (slope, intercept, r) = stats.probplot(rvs, dist=ref)
    return (qt, qe), (slope, intercept, r)
```

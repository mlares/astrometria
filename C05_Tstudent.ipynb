```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 5000
N = 10
avg = []
for _ in range(M):
    avg.append(np.mean(st.norm.rvs(0,1,N)))
```

```python
fn = st.norm(0,1./np.sqrt(N))
ft = st.t(9)
x = np.linspace(-2,2,100)
yn = fn.pdf(x)
yt = ft.pdf(x)
```

```python
plt.hist(avg, 100, normed=True)
plt.plot(x,yn)
plt.plot(x,yt)
plt.show()
```

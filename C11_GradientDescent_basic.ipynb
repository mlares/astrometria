### eg > Gradiente descendente

```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
%matplotlib inline
```

Definir la función:

una complicada:

```python
func = lambda theta: np.sin(.5*theta[0]**2-.25*theta[1])*np.cos(2*theta[0]+1-np.exp(theta[1]))
dfunc = lambda theta: [np.cos(.5*theta[0]**2-.25*theta[1])*theta[0]-2*np.sin(2*theta[0]+1-np.exp(theta[1])),
                       .25*np.cos(.5*theta[0]**2-.25*theta[1])-np.exp(theta[1])*np.sin(2*theta[0]+1-2.*np.exp(theta[1]))]
```

una fácil:

```python
func = lambda t: t[0]**2 + t[1]**2
dfunc = lambda t: [2*t[0], 2*t[1]]
```

Construir un muestreo en grid para visualizar la función

```python
N = 100
mx = 2.5
_X = np.linspace(-mx,mx,N)
_Y = np.linspace(-mx,mx,N)
_Z = np.zeros((N,N))

for ix,x in enumerate(_X):
    for iy,y in enumerate(_Y):
        _Z[iy,ix] = func([x,y])
```

```python
fig = plt.figure(figsize=(8,8))
plt.imshow(_Z)
```

```python
fig = plt.figure(figsize=(8,8))
plt.imshow(_Z, extent=[-mx, mx, -mx, mx])
```

```python
fig = plt.figure(figsize=(15,8))
plt.contourf(_X,_Y,_Z,100)
plt.colorbar()
plt.show()
```

#### Sortear un punto de inicio

```python
fig = plt.figure(figsize=(15,8))
plt.contourf(_X,_Y,_Z,100)
plt.colorbar()
plt.plot(Theta[0], Theta[1],"x",c="k", markersize=30)
plt.show()
```

#### **Implementar gradiente descendente**:

```python
Theta = np.random.rand(2)*4-2  # <-- sorteamos de nuevo el punto de inicio
```

```python
Theta
```

```python
alpha = 0.25  # <-- learning rate
M = 100
tlist = np.zeros((M,2))
for i in range(M):
    Theta = Theta - [alpha*dfunc(Theta)[0], alpha*dfunc(Theta)[1]]
    tlist[i] = Theta
```

```python
fig = plt.figure(figsize=(15,8))
plt.contourf(_X,_Y,_Z,100)
plt.colorbar()
plt.plot(tlist[:,0], tlist[:,1],"x",c="red", markersize=16)
plt.show()
```
